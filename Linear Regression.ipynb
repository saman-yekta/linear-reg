```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy import stats

page_speed = np.random.normal(10,1,10000)
purchase_amount =page_speed + np.random.normal(5,0.2,10000)
```

```python
plt.scatter(page_speed,purchase_amount)
```

```python
slope, intercept, r_value, _, _ = stats.linregress(page_speed,purchase_amount)
```

```python
def func(x):
    return slope * x + intercept

func_line = func(page_speed) 
plt.scatter(page_speed,purchase_amount)
plt.plot(page_speed,func_line,c='r')
plt.show()
```
